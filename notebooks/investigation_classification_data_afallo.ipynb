```python
import nibabel as nib

path = "../raw_data/classification_data/3D_volumetric_imaging/Brain_FMRI/"
# Charger le fichier compressé directement
img = nib.load(path+'brain_vector.nii')

# Extraire les données en tableau NumPy
data = img.get_fdata()

# Afficher la forme et quelques stats
print("Shape:", data.shape)
print("Min:", data.min(), "Max:", data.max(), "Mean:", data.mean())
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_slice(data, axis=2, index=None):
    if index is None:
        index = data.shape[axis] // 2
    slice_img = data.take(index, axis=axis)

    # Cas image couleur (H, W, 3)
    if slice_img.ndim == 3 and slice_img.shape[-1] in (3, 4):
        plt.imshow(slice_img, origin='lower')  # pas de .T, pas de cmap
    else:
        plt.imshow(slice_img.T, cmap='gray', origin='lower')

    plt.title(f"Coupe axis={axis} index={index}")
    plt.axis('off')
    plt.show()
```

```python
show_slice(data)
```
